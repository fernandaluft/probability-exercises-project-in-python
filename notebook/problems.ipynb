## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import random

def simulate(steps):
    
    #desired_results = {2:0, 4:0, 6:0, 8:0, 9:0, 10:0, 11:0, 12:0}
    times = 0
    
    for _ in range(steps):
        d1 = random.randint(1,6)
        d2 = random.randint(1,6)
        if d1 + d2 > 7 or (d1 + d2) % 2 == 0:
            times += 1
            #desired_results[d1+d2] += 1

    probability = f"{times / 1000:.2%}"

    return probability

```

```python
a = simulate(1000)
a
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO
import random

def ball(steps):

    colours = list(ball_box.values())
    outcome = {'White': 0, 'Red': 0, 'Green': 0}
    desired_outcome = 0
    whites = 0
    reds = 0
    greens = 0

    for _ in range(steps):

        for _ in range(5):
            ball = random.choice(colours)
                    
            if ball == 'White':
                outcome['White'] += 1
            if ball == 'Red':
                outcome['Red'] += 1
            if ball == 'Green':
                outcome['Green'] += 1
        
        if outcome['White'] == 3 and outcome['Red'] == 2:
            desired_outcome +=1
        if outcome['White'] == 5:
            whites += 1
        if outcome['Red'] == 5:
            reds += 1
        if outcome['Green'] == 5:
            greens += 1

    prob_desired_outcome = f"The probability of getting 3 whites and 2 reds is: {desired_outcome / steps:.2%}."
    same_colour = f"The probability of getting 5 balls in the same coulour is: {(whites + greens + reds) / steps:.2%}."

    return  prob_desired_outcome, same_colour


'''desired_outcome1 = {'White': 3, 'Red':'2}
desired_outcome2 = {'White':5}
desired_outcome3 = {'Red':5}
desired_outcome3 = {'Green':5}'''
```
